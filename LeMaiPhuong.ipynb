```python
import pandas as pd  # Importing pandas library for data manipulation
import numpy as np  # Importing numpy library for numerical operations
import matplotlib.pyplot as plt  # Importing matplotlib library for data visualization
import seaborn as sns  # Importing seaborn library for statistical plotting
```

```python
df1 = pd.read_excel("C:/Users/84869/DATA ANALYST/DATA/data1.xlsx", sheet_name='Sheet1')  # Thay 'file1.csv' bằng đường dẫn file của bạn
df2 = pd.read_excel("C:/Users/84869/DATA ANALYST/DATA/data2.xlsx", sheet_name='Sheet1')  # Thay 'file2.csv' bằng đường dẫn file của bạn


# Nối hai dataset dựa trên các cột 'sex', 'skill', 'province'
df = pd.merge(df1, df2, on=['sex', 'skill', 'province'], how='inner')  

# Hiển thị kết quả
df.head()

```

```python
df.isnull().sum()
```

```python

df_cleaned_rows = df.dropna()

# Kiểm tra dữ liệu sau khi xóa
print(df_cleaned_rows.head())
print(df_cleaned_rows.info())
```

```python
df_cleaned_rows.isnull().sum()
```

```python
#2. thống kê và trực quan cho 3 biến : sex, skill, province (cùng nhau)

stats = df.groupby(['sex', 'skill', 'province']).size().reset_index(name='count')

# Hiển thị thống kê
print(stats)

# Bước 3: Trực quan hóa dữ liệu với biểu đồ cột
plt.figure(figsize=(14, 8))
sns.barplot(data=stats, x='skill', y='count', hue='sex',  errorbar=None)
plt.title('Thống kê Skill theo Sex và Province')
plt.xlabel('Skill')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Sex')
plt.tight_layout()
plt.show()


pivot_table = stats.pivot_table(index='skill', columns=['sex', 'province'], values='count', fill_value=0)

# Kiểm tra pivot_table
print(pivot_table)

# Hiển thị heatmap
plt.figure(figsize=(16, 10))
sns.heatmap(pivot_table, annot=True, fmt='.0f', cmap='YlGnBu', linewidths=.5, cbar_kws={'label': 'Count'})
plt.title('Heatmap của Skill, Sex và Province')
plt.xlabel('Sex, Province')
plt.ylabel('Skill')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
#thống kê và trực quan cho 3 biến : skill, province (cùng nhau)

# Bước 2: Thống kê tần suất xuất hiện của các kết hợp giữa 'skill' và 'province'
stats = df.groupby(['skill', 'province']).size().reset_index(name='count')

# Hiển thị thống kê
print(stats)

# Bước 3: Trực quan hóa dữ liệu với biểu đồ cột
plt.figure(figsize=(14, 8))
sns.barplot(data=stats, x='skill', y='count', hue='province', errorbar=None)
plt.title('Thống kê Skill theo Province')
plt.xlabel('Skill')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Province')
plt.tight_layout()
plt.show()

# Trực quan hóa dữ liệu với heatmap để thể hiện mối quan hệ phức tạp hơn
# Pivot table để chuẩn bị dữ liệu cho heatmap
pivot_table = stats.pivot_table(index='skill', columns='province', values='count', fill_value=0)

# Kiểm tra pivot_table
print(pivot_table)

# Hiển thị heatmap
plt.figure(figsize=(16, 10))
sns.heatmap(pivot_table, annot=True, fmt='.0f', cmap='YlGnBu', linewidths=.5, cbar_kws={'label': 'Count'})
plt.title('Heatmap của Skill và Province')
plt.xlabel('Province')
plt.ylabel('Skill')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

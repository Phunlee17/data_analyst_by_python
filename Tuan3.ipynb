```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt  
```

```python
df = pd.read_csv("C:/Users/84869/DATA ANALYST/DATA/Invistico_Airline.csv")
```

```python
df.head()
```

```python
# Kiá»ƒm tra dá»¯ liá»‡u vÃ  cá»™t Age

df['Age'].value_counts()

# Váº½ biá»ƒu Ä‘á»“ Ä‘áº¿m cho biáº¿n phÃ¢n loáº¡i "Age"
plt.figure(figsize=(20, 8))
sns.countplot(x=df['Age'])

# Thiáº¿t láº­p tiÃªu Ä‘á» vÃ  nhÃ£n
plt.title('Biá»ƒu Ä‘á»“ Ä‘áº¿m cho biáº¿n phÃ¢n loáº¡i "Age"')
plt.xlabel('NhÃ³m tuá»•i')
plt.ylabel('Sá»‘ lÆ°á»£ng')

# Hiá»ƒn thá»‹ biá»ƒu Ä‘á»“
plt.show()
```

```python
# Váº½ biá»ƒu Ä‘á»“ heatmap
plt.figure(figsize=(8, 6))
heatmap_data = pd.crosstab(df['Seat comfort'], df['Departure/Arrival time convenient'])
sns.heatmap(heatmap_data, annot=True, cmap='Blues', cbar=True)

plt.title('Heatmap of Seat Comfort vs. Departure/Arrival Time Convenient')
plt.xlabel('Departure/Arrival Time Convenient')
plt.ylabel('Seat Comfort')
plt.show()
```

```python
# Váº½ biá»ƒu Ä‘á»“ Histogram cho biáº¿n Age
plt.figure(figsize=(10, 5))
plt.hist(df['Age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Váº½ Boxplot Ä‘á»ƒ so sÃ¡nh phÃ¢n bá»‘ Age theo Gender
plt.figure(figsize=(12, 6))
sns.boxplot(x='Age', y='Gender', data=df, palette='pastel')
plt.title('Boxplot of Gender by Age')
plt.xlabel('Age')
plt.ylabel('Gender')
plt.grid(True)
plt.show()
```

```python
# Váº½ Violin plot Ä‘á»ƒ so sÃ¡nh phÃ¢n bá»‘ Age theo Gender
plt.figure(figsize=(12, 6))
sns.violinplot(x='Age', y='Gender', data=df, palette='muted')
plt.title('Violin Plot of Gender by Age')
plt.xlabel('Age')
plt.ylabel('Gender ')
plt.grid(True)
plt.show()
```

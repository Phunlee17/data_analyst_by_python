```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt  
```

```python
df = pd.read_csv("C:/Users/84869/DATA ANALYST/DATA/Invistico_Airline.csv")
```

```python
df.head()
```

```python
# Kiểm tra dữ liệu và cột Age

df['Age'].value_counts()

# Vẽ biểu đồ đếm cho biến phân loại "Age"
plt.figure(figsize=(20, 8))
sns.countplot(x=df['Age'])

# Thiết lập tiêu đề và nhãn
plt.title('Biểu đồ đếm cho biến phân loại "Age"')
plt.xlabel('Nhóm tuổi')
plt.ylabel('Số lượng')

# Hiển thị biểu đồ
plt.show()
```

```python
# Vẽ biểu đồ heatmap
plt.figure(figsize=(8, 6))
heatmap_data = pd.crosstab(df['Seat comfort'], df['Departure/Arrival time convenient'])
sns.heatmap(heatmap_data, annot=True, cmap='Blues', cbar=True)

plt.title('Heatmap of Seat Comfort vs. Departure/Arrival Time Convenient')
plt.xlabel('Departure/Arrival Time Convenient')
plt.ylabel('Seat Comfort')
plt.show()
```

```python
# Vẽ biểu đồ Histogram cho biến Age
plt.figure(figsize=(10, 5))
plt.hist(df['Age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Vẽ Boxplot để so sánh phân bố Age theo Gender
plt.figure(figsize=(12, 6))
sns.boxplot(x='Age', y='Gender', data=df, palette='pastel')
plt.title('Boxplot of Gender by Age')
plt.xlabel('Age')
plt.ylabel('Gender')
plt.grid(True)
plt.show()
```

```python
# Vẽ Violin plot để so sánh phân bố Age theo Gender
plt.figure(figsize=(12, 6))
sns.violinplot(x='Age', y='Gender', data=df, palette='muted')
plt.title('Violin Plot of Gender by Age')
plt.xlabel('Age')
plt.ylabel('Gender ')
plt.grid(True)
plt.show()
```

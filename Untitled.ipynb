```python
import nltk
import random
from nltk.corpus import movie_reviews
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

nltk.download('movie_reviews')
nltk.download('punkt')

# Tải dữ liệu từ movie_reviews của nltk
documents = [(list(movie_reviews.words(fileid)), category)
             for category in movie_reviews.categories()
             for fileid in movie_reviews.fileids(category)]

# Shuffle dữ liệu
random.shuffle(documents)

# Tạo danh sách các câu và nhãn tương ứng
sentences = [" ".join(doc) for doc, category in documents]
labels = [category for doc, category in documents]

# Chia dữ liệu thành tập huấn luyện và tập kiểm tra
X_train, X_test, y_train, y_test = train_test_split(sentences, labels, test_size=0.2, random_state=42)

# Vector hóa văn bản
vectorizer = CountVectorizer()
X_train_counts = vectorizer.fit_transform(X_train)
X_test_counts = vectorizer.transform(X_test)

# Huấn luyện mô hình Naive Bayes
clf = MultinomialNB()
clf.fit(X_train_counts, y_train)

# Dự đoán và đánh giá mô hình
y_pred = clf.predict(X_test_counts)
print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))

# Trực quan hóa kết quả
conf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['neg', 'pos'], yticklabels=['neg', 'pos'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Câu mới cần dự đoán
new_sentence = "These days, for me, are so tiring, I don't want to live anymore."
new_sentence_counts = vectorizer.transform([new_sentence])
new_pred = clf.predict(new_sentence_counts)

# Trực quan hóa kết quả
sentiment = ['Negative' if new_pred[0] == 'neg' else 'Positive']
labels = ['Negative', 'Positive']
sizes = [1 if sentiment[0] == 'Negative' else 0, 1 if sentiment[0] == 'Positive' else 0]
colors = ['red', 'green']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Sentiment Analysis for New Sentence')
plt.show()
```

```python
import nltk
import matplotlib.pyplot as plt
import seaborn as sns
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Tải gói dữ liệu VADER nếu chưa có
nltk.download('vader_lexicon')
nltk.download('punkt')

# Câu cần phân tích
sentence = "These days, for me, are so tiring, I don't want to live anymore."

# Phân tách câu thành các từ riêng lẻ
words = nltk.word_tokenize(sentence)

# Sử dụng VADER để phân tích ngữ cảm của từng từ
sia = SentimentIntensityAnalyzer()
word_sentiments = {word: sia.polarity_scores(word)['compound'] for word in words}

# Chuẩn bị dữ liệu cho biểu đồ cột
labels = list(word_sentiments.keys())
scores = list(word_sentiments.values())
colors = ['red' if score < 0 else 'green' for score in scores]

# Tạo biểu đồ cột
plt.figure(figsize=(12, 6))
sns.barplot(x=labels, y=scores, palette=colors)
plt.axhline(0, color='grey', linewidth=0.8)
plt.title('Sentiment Analysis of Words in the Sentence')
plt.xlabel('Words')
plt.ylabel('Sentiment Score')
plt.xticks(rotation=45)
plt.show()
```

```python
# Xác định những cột visualize
columns_to_visualize = ['satisfaction', 'Gender', 'Customer Type', 'Type of Travel', 'Class']

# Tạo những subplots
fig, axs = plt.subplots(len(columns_to_visualize), 2, figsize=(12, 10))
df_in_delay = df[df['Departure Delay in Minutes'] > 0]

for i, col in enumerate(columns_to_visualize):
    # Lấy số lượng giá trị
    counts = df[col].value_counts()
    counts_in_delay = df_in_delay[col].value_counts()

    # Vẽ pie chart bên trái
    axs[i, 0].pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=45)
    axs[i, 0].set_title('Sự phân bố ' + col)
    axs[i, 0].axis('equal')

    # Vẽ pie chart bên phải
    axs[i, 1].pie(counts_in_delay, labels=counts_in_delay.index, autopct='%1.1f%%', startangle=45)
    axs[i, 1].set_title('Sự phân bố ' + col + ' khi hoãn giờ bay')
    axs[i, 1].axis('equal')

# Adjust the spacing between subplots
plt.tight_layout()
plt.show()
```

```python
# câu 1: 
# Xác định các yếu tố dịch vụ
service_columns = ['Seat comfort', 'Inflight wifi service', 'Food and drink', 'Inflight entertainment', 'Online support', 'Cleanliness', 'Leg room service']

# Tính trung bình điểm số theo giới tính
gender_service_mean = df.groupby('Gender')[service_columns].mean().reset_index()

# In ra bảng dữ liệu trung bình theo giới tính
print(gender_service_mean)
# Vẽ biểu đồ cột nhóm
plt.figure(figsize=(10, 6))
gender_service_mean_melted = pd.melt(gender_service_mean, id_vars='Gender', var_name='Service', value_name='Average Rating')
sns.barplot(x='Service', y='Average Rating', hue='Gender', data=gender_service_mean_melted, palette='viridis')

plt.title('Độ hài lòng của khách hàng theo giới tính và các yếu tố dịch vụ')
plt.xlabel('Các yếu tố dịch vụ')
plt.ylabel('Điểm trung bình')
plt.xticks(rotation=45)
plt.legend(title='Giới tính')
plt.show()
```

Biểu đồ thể hiện mức độ hài lòng của khách hàng theo giới tính (nam và nữ) với các yếu tố dịch vụ như sự thoải mái của ghế, wifi trên máy bay, thực phẩm, giải trí, hỗ trợ trực tuyến, và sạch sẽ. Cả nam và nữ đều hài lòng chung, nhưng nữ có điểm cao hơn ở "Seat comfort" và "Cleanliness," trong khi nam đánh giá cao hơn "Inflight entertainment." Các yếu tố như "Inflight wifi service" và "Food and drink" có điểm số thấp, cho thấy cần cải thiện. Biểu đồ giúp nhận diện sự khác biệt và những lĩnh vực cần nâng cao chất lượng dịch vụ.

```python
# Câu 2:
# Tạo bảng tổng hợp cho khoảng cách bay trung bình theo dịch vụ hỗ trợ trực tuyến và loại chuyến đi
pivot_table_online_support = df.pivot_table(index='Online support', columns='Type of Travel', values='Flight Distance', aggfunc='mean')

# In bảng dữ liệu
print(pivot_table_online_support)
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table_online_support, annot=True, cmap='YlOrBr', fmt='.1f')

plt.title('Khoảng cách bay trung bình tương ứng với điểm đánh giá hỗ trợ trực tuyến và mục đích bay')
plt.xlabel('Mục đích bay')
plt.ylabel('Điểm đánh giá dịch vụ hỗ trợ trực tuyến')
plt.show()
```

Biểu đồ cho thấy điểm đánh giá dịch vụ hỗ trợ trực tuyến cho du lịch công tác là khoảng 2078.6, trong khi điểm cho du lịch cá nhân là khoảng 1677.0. Điều này cho thấy dịch vụ hỗ trợ trực tuyến được đánh giá cao hơn đối với khách hàng đi công tác so với khách hàng đi du lịch cá nhân.

Sau đó, độ hài lòng giảm dần và trở nên không ổn định từ 50 đến 80 tuổi. Phần bóng mờ xung quanh đường biểu diễn phản ánh khoảng tin cậy của dữ liệu, cho thấy mức độ biến động. Nhìn chung, khách hàng trẻ (dưới 40 tuổi) có xu hướng hài lòng cao hơn, trong khi khách hàng lớn tuổi hơn trải qua sự giảm sút về độ hài lòng. Biểu đồ này chỉ ra rằng cần điều chỉnh trải nghiệm dịch vụ theo độ tuổi để nâng cao sự hài lòng của khách hàng.

```python
# Câu 4: 
# Tính số lượng hài lòng theo loại chuyến đi
satisfaction_count = df.groupby(['Type of Travel', 'satisfaction']).size().reset_index(name='Count')

# Vẽ biểu đồ cột
plt.figure(figsize=(8, 5))
sns.barplot(data=satisfaction_count, x='Type of Travel', y='Count', hue='satisfaction', palette='viridis')

plt.title('Độ hài lòng của khách hàng theo loại chuyến đi')
plt.xlabel('Loại chuyến đi')
plt.ylabel('Số lượng')
plt.grid(axis='y')
plt.legend(title='Độ hài lòng')
plt.show()
```

Biểu đồ thể hiện độ hài lòng của khách hàng theo loại chuyến đi (du lịch công tác và du lịch cá nhân) cho thấy sự khác biệt rõ rệt. Đối với du lịch công tác, số lượng khách hàng hài lòng (satisfied) đạt khoảng 50,000 người, trong khi số lượng không hài lòng (dissatisfied) chỉ khoảng 30,000 người. Ngược lại, trong nhóm khách hàng du lịch cá nhân, số lượng hài lòng chỉ khoảng 20,000, trong khi số lượng không hài lòng lên đến khoảng 30,000 người. Kết quả này cho thấy khách hàng đi công tác có xu hướng hài lòng hơn so với khách hàng đi du lịch cá nhân. Điều này có thể phản ánh sự khác biệt trong kỳ vọng và trải nghiệm dịch vụ giữa hai nhóm. Biểu đồ cung cấp cái nhìn rõ ràng về sự phân bố độ hài lòng, chỉ ra rằng cần cải thiện trải nghiệm cho khách hàng du lịch cá nhân.
